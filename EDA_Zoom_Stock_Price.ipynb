```python
#importing datetime module from datetime library to use "resample feature"
from datetime import datetime
```

```python
#represented below are the max values for end of year frequency
df_zoom.resample(rule="A").max()["High"].plot()
```

```python
#Quarterly Start Data Frequency
df_zoom.resample(rule="QS").max()["High"].plot()
```

```python
#Business Year End Data Frequency
df_zoom.resample(rule="BA").max()
```

```python
#Business Quarterly Summary Freqeuncy
df_zoom.resample(rule="BQS").max()
```

```python
#Bar Graph for Business Quarter Frequency for Stock Price "High"
df_zoom["High"].resample(rule='BQS').mean().plot(kind="bar")
```

```python
#Graph of Monthly Mean Stock Highs
df_zoom["High"].resample(rule="M").mean().plot(kind="bar", figsize= (14,7))
```

```python
#displaying the Zoom Stock High Rolling Mean for 10 days and the first 30 rows of the data set
df_zoom['High'].rolling(10).mean().head(30)
```

```python
#adding a new column to data frame called "High: 10 Days Rolling Mean" 
df_zoom['High: 10 Days Rolling Mean'] = df_zoom['High'].rolling(10).mean()
```

```python
#displaying the first 30 row values with new column included
df_zoom.head(30)
```

```python
#graphs comparison of Series Stock High Prices with Series Stock Rolling Mean with mean value of 10
df_zoom[["High","High: 10 Days Rolling Mean"]].plot(figsize=(12,4))
```
